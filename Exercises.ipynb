```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1]) 
ax2 = fig.add_axes([0.2,.5,.2,.2])

ax1.plot(x,y, color='r')
ax2.plot(x,y)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.set_title('hola')

ax2.plot(x,y)
ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])


```

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y,ls='--',lw=3)
axes[1].plot(x,z,color='red',lw=3)
```
