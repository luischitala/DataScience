```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import pandas as pd
```

MLE como base para la regresión logística, considerando 
el problema de MLE 

donde
y; clase o categoria de cada elemento y x; son los atributos de
cada elemento, donde además cada elemento del dataset satisface
una distribución de bernoulli

```python
def likelihood(y, yp):
    return yp*y+(1-yp)*(1-y)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

Y = np.arange(0,1,0.01)
YP = np.arange(0,1,0.01)
Y, YP = np.meshgrid(Y,YP)
Z = likelihood(Y,YP)
surf = ax.plot_surface(Y,YP,Z, cmap=cm.coolwarm)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
#La función de verosimilitud tiene datos en donde los quisimos
#Da probabilidades altas cuando las predicciones del modelo
#coinciden con las categorias del conjunto de datos.
```
