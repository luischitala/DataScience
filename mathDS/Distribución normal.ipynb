```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Distribución normal teórica

u media de la distribución
σ = desviación estandar de la distribución 

```python
#mu promedio, sigma desvición estandar
def gaussian(x, mu, sigma):
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*pow((x-mu)/sigma,2))
```

```python
x = np.arange(-4,4, 0.1)
#La desviación estandar engorda o adelgaza
y = gaussian(x, 0.0, 0.5)

plt.plot(x,y)
```

```python
dist = norm(0, 1)
x = np.arange(-4,4,0.1)
y = [dist.pdf(value) for value in x]
plt.plot(x,y)
```

```python
dist = norm(0,1)
x = np.arange(-4,4,0.1)
#Distribución de probabilidad acumulada
y = [dist.cdf(value) for value in x]
plt.plot(x,y)
```

Distribución Normal(Gausiana) a partir de los datos

```python
df = pd.read_excel('s057.xls')
arr = df['Normally Distributed Housefly Wing Lengths'].values[4:]
values, dist = np.unique(arr, return_counts=True)
plt.bar(values,dist)
```

```python
#Estimación de una distribución o paramétrica
mu = arr.mean()
sigma = arr.std()
x = np.arange(30,60, 0.1)
dist = norm(mu, sigma)
y = [dist.pdf(value)for value in x]
plt.plot(x,y)
values, dist = np.unique(arr, return_counts=True)
plt.bar(values,dist/len(arr))
```
